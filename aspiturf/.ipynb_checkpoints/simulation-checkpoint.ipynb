```python
MAX_HORSES = 19

cols = [
#  'id',
 'comp',
 'jour',
 'hippo',
#  'numcourse',
 'cl',
 'dist',
#  'partant',
#  'typec',
#  'cheque',
 'numero',
 'cheval',
 'sexe',
 'age',
 'cotedirect',
 'coteprob',
 'recence',
#  'ecurie',
#  'ecar',
#  'redkm',
#  'redkmInt',
 'corde',
 'defoeil',
 'defoeilPrec',
 'recul',
#  'musiqueche',
 'm1',
 'm2',
 'm3',
 'm4',
 'm5',
 'm6',
 'jockey',
 'musiquejoc',
 'montesdujockeyjour',
 'couruejockeyjour',
#  'victoirejockeyjour',
 'entraineur',
 'musiqueent',
 'monteentraineurjour',
 'courueentraineurjour',
#  'victoireentraineurjour',
 'coursescheval',
 'victoirescheval',
 'placescheval',
 'coursesentraineur',
 'victoiresentraineur',
 'placeentraineur',
 'coursesjockey',
 'victoiresjockey',
 'placejockey',
#  'dernierhippo',
#  'dernierealloc',
#  'derniernbpartants',
#  'dernieredist',
#  'derniereplace',
#  'dernierecote',
#  'dernierJoc',
#  'dernierEnt',
#  'dernierProp',
#  'proprietaire',
 'poidmont',
 'pourcVictJock',
 'pourcPlaceJock',
 'pourcVictCheval',
 'pourcPlaceCheval',
 'pourcVictEnt',
 'pourcPlaceEnt',
 'pourcVictEntHippo',
 'pourcVictJockHippo',
 'pourcPlaceEntHippo',
 'pourcPlaceJockHippo',
 'pourcVictChevalHippo',
 'pourcPlaceChevalHippo',
 'nbrCourseJockHippo',
 'nbrCourseEntHippo',
 'nbrCourseChevalHippo',
 'nbCourseCouple',
 'nbVictCouple',
 'nbPlaceCouple',
 'TxVictCouple',
 'TxPlaceCouple',
 'nbCourseCoupleHippo',
 'nbVictCoupleHippo',
 'nbPlaceCoupleHippo',
 'TxVictCoupleHippo',
 'TxPlaceCoupleHippo',
#  'pere',
#  'mere',
 'gainsCarriere',
 'gainsVictoires',
 'gainsPlace',
 'gainsAnneeEnCours',
 'gainsAnneePrecedente',
#  'engagement',
 'handicapDistance',
 'handicapPoids',
#  'indicateurInedit',
#  'tempstot',
#  'recordG', ## TIME_TO_SEC(TIME(REPLACE(CONCAT('00:0',LEFT(recordG,4)),\"'\",':'))) as recordGint
#  'txreclam',
#  'createdat',
#  'updatedat'
]
data = pd.read_sql("SELECT {} FROM cachedate WHERE hippo is not NULL and cotedirect > 0 AND partant < {} AND corde > 0 AND typec = 'Plat' AND YEAR(jour) in (2016,2017,2018) AND cl in ({})".format(','.join(cols),MAX_HORSES,','.join(["'1er'"]+["'{}e'".format(n) for n in range(40)])), engine)
```

```python
data = data.join(data['musiquejoc'].apply(lambda x: ' '.join(re.sub(r'([^\d ])','',re.sub(r'\(.*\)','',x)).split()[:5])).str.split(expand=True).add_prefix('jocm_'))
data = data.join(data['musiqueent'].apply(lambda x: ' '.join(re.sub(r'([^\d ])','',re.sub(r'\(.*\)','',x)).split()[:5])).str.split(expand=True).add_prefix('entm_'))

data['classement'] = data.apply(lambda x: int(x['cl'].replace('er','').replace('e','')), axis=1)

data['med'] = data.apply(lambda x: np.median([x['m1'],x['m2'],x['m3'],x['m4'],x['m5'],x['m6']]), axis=1)
data['mea'] = data.apply(lambda x: np.mean([x['m1'],x['m2'],x['m3'],x['m4'],x['m5'],x['m6']]), axis=1)

data['joc_mea'] = data['musiquejoc'].apply(lambda x: np.mean([int(v) for v in re.sub(r'([^\d ])','',re.sub(r'\(.*\)','',x)).split()[:5]])).fillna(-999)
data['joc_med'] = data['musiquejoc'].apply(lambda x: np.median([int(v) for v in re.sub(r'([^\d ])','',re.sub(r'\(.*\)','',x)).split()[:5]])).fillna(-999)

data['ent_mea'] = data['musiqueent'].apply(lambda x: np.mean([int(v) for v in re.sub(r'([^\d ])','',re.sub(r'\(.*\)','',x)).split()[:5]])).fillna(-999)
data['ent_med'] = data['musiqueent'].apply(lambda x: np.median([int(v) for v in re.sub(r'([^\d ])','',re.sub(r'\(.*\)','',x)).split()[:5]])).fillna(-999)

data = data.fillna(-999)

data['jour'] = pd.to_datetime(data['jour'])

data = data.astype({'numero':int,**{f'jocm_{n}':int for n in range(5)},**{f'entm_{n}':int for n in range(5)}})
data.head(5)
```

```python
competition = data.loc[data.corde==1]


# competition.columns = [(c+'_1') if c != 'comp' else c for c in competition.columns]
for n in range(2,19):
    competition = competition.merge(data.loc[data.corde==n], how='outer', on=['comp','jour','hippo'], suffixes=['',f'_{n}'])
competition = competition.fillna(-999)

info = pd.read_sql("SELECT comp,meteo,sex,temperature,forceVent,directionVent,nebulositeLibelleCourt FROM caractrap",con=engine)

competition = competition.merge(info, on='comp')

competition.head(5)
```

```python
to_remove = []
def find_winner(x):
    if x['classement'] == 1:
        return int(1)
    for n in range(2,MAX_HORSES):
        if x[f'classement_{n}'] == 1:
            return int(n)
    to_remove.append(x.name)

def find_favorite(x, cotetype):
    if x[cotetype] != -999:
        min_cote = x[cotetype]
        fav = 1
    else:
        min_cote = float('Infinity')
        fav = -1
    for n in range(2,MAX_HORSES):
        cote = x[f'{cotetype}_{n}']
        if cote != -999:
            if cote < min_cote:
                min_cote = cote
                fav = n
    return fav
    
def find_favorite_cote(x, cotetype):
    if x[cotetype] != -999:
        min_cote = x[cotetype]
        fav = 1
    else:
        min_cote = float('Infinity')
        fav = -1
    for n in range(2,MAX_HORSES):
        cote = x[f'{cotetype}_{n}']
        if cote != -999:
            if cote < min_cote:
                min_cote = cote
                fav = n
    return min_cote
    
def find_prob(x):
    return find_favorite(x,'coteprob')

def find_prob_cote(x):
    return find_favorite_cote(x,'coteprob')

def find_direct(x):
    return find_favorite(x,'cotedirect')

def find_direct_cote(x):
    return find_favorite_cote(x,'cotedirect')

competition.apply(find_winner, axis=1)

clean = competition.drop(to_remove)

y = pd.DataFrame({
    'cl':clean.apply(find_winner, axis=1),
    'prob':clean.apply(find_prob, axis=1),
    'direct':clean.apply(find_direct, axis=1),
    'prob_cote':clean.apply(find_prob_cote, axis=1),
    'direct_cote':clean.apply(find_direct_cote, axis=1),
})
print(to_remove)
y.head(5)
```

```python
X_raw = clean.drop(
    [
        'comp',
        'jour',
#         'cotedirect',
#         'coteprob',
        'cl',
        'classement'
    ]
    +[f'cl_{n}' for n in range(2,19)]
    +[f'classement_{n}' for n in range(2,19)]
#     +[f'cotedirect_{n}' for n in range(2,19)]
#     +[f'coteprob_{n}' for n in range(2,19)]
    
    , axis=1)
cols = u_col.types(X_raw, False)
C = cols['text']+cols['tags']+cols['cats']

X = pd.concat([
#         u_cat.tfidf(X_raw, tfidf),
#         u_cat.dummies(X_raw, C),
        u_cat.label_encoder(X_raw, C + cols['bool']),
#         u_cat.get_count(X_raw, "nom_produit", 'nom_magasin'),
#         u_cat.get_distinct_count(X_raw, "nom_produit", 'nom_magasin'),
        X_raw[cols['num']]
    ], axis=1).fillna(-999)

X.head()
```

```python
X_data = X[clean.jour.dt.year <= 2017]
y_data = y['cl'][clean.jour.dt.year <= 2017]
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.15, random_state=20, stratify=y_data)
```

```python
params = {
    'max_depth': 8,
    'n_estimators': 100,
    'n_jobs': 4,
#     'lambda':4,
    'gamma': 6,
#     'subsample': .9
    'eta':.05
}

clf = xgb.XGBClassifier(**params)
clf.fit(X_train, y_train,
    eval_set=[(X_train, y_train),(X_test, y_test)],
    eval_metric='mlogloss',
    verbose=5,
    early_stopping_rounds=5
)
```

```python
ax = lgb.plot_metric(clf.evals_result_)
plt.show()
```

```python
def sim(PRELEV, MAX_BET):
    cost = [0]
    curr = [0]
    rend = []
    races= []
    wins = []
    ratio = []
    vals = []
    bank = [0]
    days = pd.to_datetime(clean[clean.jour.dt.year == 2018]['jour'].unique()).sort_values()

    for day in tqdm(days):
        lines = clean.jour==day
        X_day = X[lines]

        y_pred_prob = clf.predict_proba(X_day)
        y_pred = np.argmax(y_pred_prob, axis=1)+1

        y_day = y['cl'][lines]

        N = len(y_day)
        b = bank[-1]

        if curr[-1] < N:
            missing = N - curr[-1]
            from_bank = min(missing,b)
            b -= from_bank
            missing -= from_bank

            cost.append(cost[-1]+missing)
            val = 1
        else:
            cost.append(cost[-1])
            val = min(round(curr[-1]/N,2),MAX_BET*curr[-1])

        vals.append(val)
        c = curr[-1] - val * N
        cote = []
        for pred, cl, (i,x) in zip(y_pred,y_day, X_day.iterrows()):
            if pred == cl:
                if pred == 1:
                    cote.append(x['cotedirect'])
                else:
                    cote.append(x[f'cotedirect_{pred}'])
        ratio.append(round(len(cote)/N,2))
        rend.append(sum(cote)/N)

        c += sum(cote)*val
        c = max(0,c)
        bank.append(b+c*PRELEV)
        curr.append(c*(1-PRELEV))
        races.append(N)
        wins.append(len(cote))

    return {
        'days':days,
        'bank':bank[1:],
        'cost':cost[1:],
        'curr':curr[1:],
        'vals':vals,
        'rend':rend,
        'races':races,
        'wins':wins,
        'ratio':ratio
    }
```

```python
result = []
bank = []
curr = []
cost = []
prelev = []
max_bet = []
for PRELEV in np.arange(0,.02,.01):
    for MAX_BET in np.arange(0,.1,.01):
        s = sim(PRELEV,MAX_BET)
        prelev.append(PRELEV)
        max_bet.append(MAX_BET)
        result.append(s['bank'][-1]+s['curr'][-1])
        bank.append(s['bank'][-1])
        curr.append(s['curr'][-1])
        cost.append(s['cost'][-1])
        print(PRELEV,MAX_BET,result[-1])
```

```python
df = pd.DataFrame({'prelev':prelev,'max_bet':max_bet,'bank':bank,'curr':curr,'cost':cost,'result':result})
```
